```python
# import the necessary packages
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report
from sklearn import datasets
import numpy as np
import cv2
import pandas as pd
```

```python
from nolearn.dbn import DBN
```

```python
dataset = pd.read_csv('data/dati/glass.data.csv')
```

```python
dataset.dtypes
```

```python
dataset['Type of glass'].unique()
```

```python
dataset.drop('Id number',axis=1,inplace=True)
```

The Deep Belief Network assumes that our data is scaled in the range [0, 1.0]:

```python
def normalize(col, df):
    for i in col:
        df[i] = (df[i] - np.min(df[i]))/(np.max(df[i]) - np.min(df[i]))
```

```python
normalize(dataset.columns.drop('Type of glass'), dataset)
```

```python
train_X = np.matrix(dataset.drop('Type of glass',axis=1))
train_y = np.array(dataset['Type of glass'])
```

```python
dbn = DBN([train_X.shape[1], 300, 6],learn_rates = 0.3,learn_rate_decays = 0.9,epochs = 1000,verbose = 1)
dbn.fit(train_X, train_y)
```

```python
preds = dbn.predict(train_X)
print classification_report(train_y, preds)
```
