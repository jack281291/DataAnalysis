# Get Dummies

```python
import pandas as pd
import numpy as np
```

```python
recent_grads = pd.read_csv('data/dati/recent_grads.csv')
```

```python
recent_grads.head()
```

```python
major_dummies = pd.get_dummies(recent_grads.Major, prefix='MAJOR_')
```

```python
major_dummies
```

```python
recent_grads.drop('Major', axis=1,inplace=True)
```

```python
recent_grads = recent_grads.join(major_dummies)
```
